```python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import optuna
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score

# Verificar si hay GPU disponible
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Utilizando: {device}")

# Cargar el dataset
dataset = fetch_california_housing()
input_data = dataset.data
Xtrain, Xtest, Ytrain, Ytest = train_test_split(input_data, dataset.target, test_size=0.3, random_state=0)

# Normalizar los datos
scaler = StandardScaler()
Xtrain_scaled = scaler.fit_transform(Xtrain)
Xtest_scaled = scaler.transform(Xtest)

# Convertir a tensores de PyTorch
X_train_tensor = torch.FloatTensor(Xtrain_scaled).to(device)
y_train_tensor = torch.FloatTensor(Ytrain.reshape(-1, 1)).to(device)
X_test_tensor = torch.FloatTensor(Xtest_scaled).to(device)
y_test_tensor = torch.FloatTensor(Ytest.reshape(-1, 1)).to(device)

# Crear DataLoaders
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
test_dataset = TensorDataset(X_test_tensor, y_test_tensor)

# Definir la red neuronal
class NeuralNetwork(nn.Module):
    def __init__(self, input_size, hidden_layers, neurons_per_layer, dropout_rate=0.2):
        super(NeuralNetwork, self).__init__()
        
        layers = []
        # Primera capa
        layers.append(nn.Linear(input_size, neurons_per_layer))
        layers.append(nn.ReLU())
        layers.append(nn.Dropout(dropout_rate))
        
        # Capas ocultas
        for _ in range(hidden_layers - 1):
            layers.append(nn.Linear(neurons_per_layer, neurons_per_layer))
            layers.append(nn.ReLU())
            layers.append(nn.Dropout(dropout_rate))
        
        # Capa de salida
        layers.append(nn.Linear(neurons_per_layer, 1))
        
        self.model = nn.Sequential(*layers)
    
    def forward(self, x):
        return self.model(x)

# Función para entrenar el modelo
def train_model(model, train_loader, criterion, optimizer, epochs):
    model.train()
    losses = []
    
    for epoch in range(epochs):
        epoch_loss = 0
        for X_batch, y_batch in train_loader:
            # Forward pass
            y_pred = model(X_batch)
            loss = criterion(y_pred, y_batch)
            
            # Backward pass y optimización
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            
            epoch_loss += loss.item()
            
        avg_loss = epoch_loss / len(train_loader)
        losses.append(avg_loss)
        
        if epoch % 10 == 0:
            print(f"Epoch {epoch}: Loss = {avg_loss:.6f}")
    
    return losses

# Función para evaluar el modelo
def evaluate_model(model, test_loader, criterion):
    model.eval()
    test_loss = 0
    all_preds = []
    all_targets = []
    
    with torch.no_grad():
        for X_batch, y_batch in test_loader:
            y_pred = model(X_batch)
            test_loss += criterion(y_pred, y_batch).item()
            
            all_preds.extend(y_pred.cpu().numpy())
            all_targets.extend(y_batch.cpu().numpy())
    
    avg_test_loss = test_loss / len(test_loader)
    mse = mean_squared_error(all_targets, all_preds)
    r2 = r2_score(all_targets, all_preds)
    
    return avg_test_loss, mse, r2

# Función objetivo para Optuna (mejorada)
def objective(trial):
    # Hiperparámetros a optimizar (rangos ampliados)
    hidden_layers = trial.suggest_int('hidden_layers', 1, 8)
    neurons_per_layer = trial.suggest_int('neurons_per_layer', 32, 512)
    learning_rate = trial.suggest_float('learning_rate', 1e-5, 1e-1, log=True)
    batch_size = trial.suggest_categorical('batch_size', [16, 32, 64, 128, 256])
    dropout_rate = trial.suggest_float('dropout_rate', 0.0, 0.5)
    
    # Crear DataLoader con el batch size sugerido
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    test_loader = DataLoader(test_dataset, batch_size=batch_size)
    
    # Crear y entrenar el modelo
    input_size = X_train_tensor.shape[1]
    model = NeuralNetwork(input_size, hidden_layers, neurons_per_layer, dropout_rate).to(device)
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)
    
    # Entrenamiento por más épocas para evaluar mejor cada configuración
    train_model(model, train_loader, criterion, optimizer, epochs=50)
    
    # Evaluación
    _, mse, _ = evaluate_model(model, test_loader, criterion)
    
    return mse  # Minimizar el MSE

# Ejecutar la optimización con Optuna (más pruebas)
def run_optuna_optimization(n_trials=100):  # Incrementar el número de pruebas
    study = optuna.create_study(direction='minimize')
    study.optimize(objective, n_trials=n_trials)
    
    print("Mejores hiperparámetros:")
    print(study.best_params)
    print(f"Mejor valor objetivo (MSE): {study.best_value}")
    
    # Visualizar la importancia de los parámetros
    optuna.visualization.plot_param_importances(study)
    
    return study.best_params

# Entrenar el modelo final con los mejores hiperparámetros
def train_final_model(best_params, epochs=100):
    # Configurar el modelo con los mejores hiperparámetros
    input_size = X_train_tensor.shape[1]
    model = NeuralNetwork(
        input_size, 
        best_params['hidden_layers'], 
        best_params['neurons_per_layer'], 
        best_params['dropout_rate']
    ).to(device)
    
    # DataLoaders
    train_loader = DataLoader(train_dataset, batch_size=best_params['batch_size'], shuffle=True)
    test_loader = DataLoader(test_dataset, batch_size=best_params['batch_size'])
    
    # Entrenar
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=best_params['learning_rate'])
    losses = train_model(model, train_loader, criterion, optimizer, epochs)
    
    # Evaluar
    test_loss, mse, r2 = evaluate_model(model, test_loader, criterion)
    print(f"Pérdida final en test: {test_loss:.6f}")
    print(f"Error cuadrático medio: {mse:.6f}")
    print(f"R^2 score: {r2:.6f}")
    
    # Graficar la curva de aprendizaje
    plt.figure(figsize=(10, 6))
    plt.plot(losses)
    plt.title('Curva de Aprendizaje')
    plt.xlabel('Épocas')
    plt.ylabel('Pérdida')
    plt.grid(True)
    plt.show()
    
    return model, losses, test_loss, mse, r2

# Ejecutar todo el proceso
if __name__ == "__main__":
    print("Iniciando la optimización de hiperparámetros...")
    best_params = run_optuna_optimization(n_trials=30)  # Reducir para pruebas, aumentar para resultados finales
    
    print("\nEntrenando el modelo final con los mejores hiperparámetros...")
    final_model, losses, test_loss, mse, r2 = train_final_model(best_params, epochs=150)
    
    # Hacer algunas predicciones
    sample_indices = np.random.choice(len(Xtest), 10)
    X_samples = X_test_tensor[sample_indices]
    y_true = y_test_tensor[sample_indices].cpu().numpy()
    
    final_model.eval()
    with torch.no_grad():
        y_pred = final_model(X_samples).cpu().numpy()
    
    print("\nComparativa de predicciones vs valores reales:")
    for i, (true, pred) in enumerate(zip(y_true, y_pred)):
        print(f"Muestra {i+1}: Real = {true[0]:.4f}, Predicción = {pred[0]:.4f}, Diff = {abs(true[0]-pred[0]):.4f}")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score
import math
from sklearn.model_selection import KFold

# Configuración de semilla para reproducibilidad
torch.manual_seed(42)
np.random.seed(42)

# Verificar disponibilidad de GPU
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Usando: {device}")

# Cargar dataset
dataset = datasets.fetch_california_housing()
print(f"MIN = {np.min(dataset.target)}, MAX = {np.max(dataset.target)}")

# Técnicas avanzadas de preprocesamiento
# 1. Normalización de datos
scaler_X = StandardScaler()
X_scaled = scaler_X.fit_transform(dataset.data)

# Escalado robusto para la variable objetivo
scaler_y = StandardScaler()
y_reshaped = dataset.target.reshape(-1, 1)
y_scaled = scaler_y.fit_transform(y_reshaped).flatten()

# 2. Gestión de valores atípicos (opcional)
# Eliminamos valores extremos en la variable objetivo
z_scores = np.abs((y_scaled - np.mean(y_scaled)) / np.std(y_scaled))
non_outliers = z_scores < 3  # Mantener datos dentro de 3 desviaciones estándar
X_clean = X_scaled[non_outliers]
y_clean = y_scaled[non_outliers]

# Dividir los datos (80% entrenamiento, 20% prueba)
X_train, X_test, y_train, y_test = train_test_split(
    X_clean, y_clean, test_size=0.2, random_state=42
)

# Convertir a tensores de PyTorch
X_train_tensor = torch.FloatTensor(X_train).to(device)
y_train_tensor = torch.FloatTensor(y_train).to(device)
X_test_tensor = torch.FloatTensor(X_test).to(device)
y_test_tensor = torch.FloatTensor(y_test).to(device)

# Modelo mejorado basado en los mejores hiperparámetros encontrados
class ImprovedHousingNN(nn.Module):
    def __init__(self, input_size, hidden_layers=3, neurons_per_layer=175, dropout_rate=0.1):
        super(ImprovedHousingNN, self).__init__()
        
        layers = []
        # Primera capa
        layers.append(nn.Linear(input_size, neurons_per_layer))
        layers.append(nn.ReLU())
        layers.append(nn.BatchNorm1d(neurons_per_layer))
        layers.append(nn.Dropout(dropout_rate))
        
        # Capas ocultas
        for _ in range(hidden_layers - 1):
            layers.append(nn.Linear(neurons_per_layer, neurons_per_layer))
            layers.append(nn.ReLU())
            layers.append(nn.BatchNorm1d(neurons_per_layer))
            layers.append(nn.Dropout(dropout_rate))
        
        # Capa de salida
        layers.append(nn.Linear(neurons_per_layer, 1))
        
        self.model = nn.Sequential(*layers)
        
        # Inicialización de pesos mejorada
        self._initialize_weights()
    
    def _initialize_weights(self):
        for m in self.modules():
            if isinstance(m, nn.Linear):
                # Inicialización He (para ReLU)
                nn.init.kaiming_normal_(m.weight, mode='fan_in', nonlinearity='relu')
                if m.bias is not None:
                    nn.init.zeros_(m.bias)
    
    def forward(self, x):
        return self.model(x).squeeze()

# Configuraciones mejoradas basadas en Optuna
batch_size = 16  # Mejor resultado de Optuna
hidden_layers = 3
neurons_per_layer = 175
learning_rate = 0.00046
dropout_rate = 0.10

# Crear DataLoader
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

# Cross-validation para mejorar la generalización
k_folds = 5
kfold = KFold(n_splits=k_folds, shuffle=True, random_state=42)

# Listas para almacenar resultados de cada fold
fold_results = []

# Definir función para entrenar y evaluar el modelo
def train_and_evaluate(model, train_loader, val_X, val_y, epochs=200, early_stop_patience=20):
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=learning_rate, weight_decay=1e-5)
    
    # Learning rate scheduler
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(
        optimizer, mode='min', factor=0.5, patience=10, verbose=False
    )
    
    train_losses = []
    val_losses = []
    best_val_loss = float('inf')
    counter = 0
    best_model_state = None
    
    for epoch in range(epochs):
        model.train()
        running_loss = 0.0
        
        for inputs, targets in train_loader:
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, targets)
            loss.backward()
            
            # Gradient clipping para estabilidad
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            
            optimizer.step()
            running_loss += loss.item()
        
        avg_train_loss = running_loss / len(train_loader)
        train_losses.append(avg_train_loss)
        
        # Evaluación en validación
        model.eval()
        with torch.no_grad():
            val_outputs = model(val_X)
            val_loss = criterion(val_outputs, val_y)
            val_rmse = torch.sqrt(val_loss).item()
        
        val_losses.append(val_rmse)
        scheduler.step(val_rmse)
        
        # Imprimir progreso cada 20 épocas
        if (epoch + 1) % 20 == 0:
            print(f'Época {epoch+1}/{epochs}, Pérdida: {avg_train_loss:.6f}, Val RMSE: {val_rmse:.6f}')
        
        # Guardar mejor modelo
        if val_rmse < best_val_loss:
            best_val_loss = val_rmse
            best_model_state = model.state_dict().copy()
            counter = 0
        else:
            counter += 1
            if counter >= early_stop_patience:
                print(f'Early stopping en época {epoch+1}')
                break
    
    # Cargar el mejor modelo
    model.load_state_dict(best_model_state)
    return model, best_val_loss, train_losses, val_losses

# Función para validación cruzada
def run_cross_validation(X, y):
    cv_results = []
    
    for fold, (train_idx, val_idx) in enumerate(kfold.split(X)):
        print(f"\nEntrenando fold {fold+1}/{k_folds}")
        
        # Preparar datos para este fold
        X_fold_train = X[train_idx]
        y_fold_train = y[train_idx]
        X_fold_val = X[val_idx]
        y_fold_val = y[val_idx]
        
        # Convertir a tensores
        X_fold_train_tensor = torch.FloatTensor(X_fold_train).to(device)
        y_fold_train_tensor = torch.FloatTensor(y_fold_train).to(device)
        X_fold_val_tensor = torch.FloatTensor(X_fold_val).to(device)
        y_fold_val_tensor = torch.FloatTensor(y_fold_val).to(device)
        
        # Crear DataLoader
        fold_train_dataset = TensorDataset(X_fold_train_tensor, y_fold_train_tensor)
        fold_train_loader = DataLoader(fold_train_dataset, batch_size=batch_size, shuffle=True)
        
        # Crear y entrenar modelo
        fold_model = ImprovedHousingNN(
            input_size=X.shape[1], 
            hidden_layers=hidden_layers,
            neurons_per_layer=neurons_per_layer, 
            dropout_rate=dropout_rate
        ).to(device)
        
        fold_model, val_loss, _, _ = train_and_evaluate(
            fold_model, fold_train_loader, X_fold_val_tensor, y_fold_val_tensor
        )
        
        cv_results.append(val_loss)
        print(f"Fold {fold+1} RMSE: {val_loss:.6f}")
    
    return cv_results

# Ejecutar validación cruzada
print("Ejecutando validación cruzada...")
cv_rmses = run_cross_validation(X_clean, y_clean)
print(f"RMSE promedio en CV: {np.mean(cv_rmses):.6f} ± {np.std(cv_rmses):.6f}")

# Entrenar modelo final con todos los datos
print("\nEntrenando modelo final con todos los datos...")
final_model = ImprovedHousingNN(
    input_size=X_clean.shape[1], 
    hidden_layers=hidden_layers,
    neurons_per_layer=neurons_per_layer, 
    dropout_rate=dropout_rate
).to(device)

train_dataset_full = TensorDataset(X_train_tensor, y_train_tensor)
train_loader_full = DataLoader(train_dataset_full, batch_size=batch_size, shuffle=True)

final_model, _, train_losses, val_losses = train_and_evaluate(
    final_model, train_loader_full, X_test_tensor, y_test_tensor, epochs=300
)

# Guardar el modelo
torch.save(final_model.state_dict(), 'best_california_housing_model.pth')

# Evaluación final
final_model.eval()
with torch.no_grad():
    # Predicciones en datos escalados
    y_pred_scaled = final_model(X_test_tensor).cpu().numpy()
    
    # Calcular métricas en escala normalizada
    scaled_mse = mean_squared_error(y_test, y_pred_scaled)
    scaled_rmse = np.sqrt(scaled_mse)
    scaled_r2 = r2_score(y_test, y_pred_scaled)
    
    # Desnormalizar predicciones y valores reales
    y_test_original = scaler_y.inverse_transform(y_test.reshape(-1, 1)).flatten()
    y_pred_original = scaler_y.inverse_transform(y_pred_scaled.reshape(-1, 1)).flatten()
    
    # Calcular métricas en escala original
    unscaled_mse = mean_squared_error(y_test_original, y_pred_original)
    unscaled_rmse = np.sqrt(unscaled_mse)
    unscaled_r2 = r2_score(y_test_original, y_pred_original)

print(f"\nResultados finales:")
print(f"RMSE escalado: {scaled_rmse:.6f}")
print(f"RMSE no escalado: {unscaled_rmse:.6f}")
print(f"R² escalado: {scaled_r2:.6f}")
print(f"R² no escalado: {unscaled_r2:.6f}")

# Visualizaciones
plt.figure(figsize=(16, 6))

# Gráfico de pérdida durante el entrenamiento
plt.subplot(1, 3, 1)
plt.plot(train_losses, label='Pérdida entrenamiento')
plt.title('Pérdida durante entrenamiento')
plt.xlabel('Épocas')
plt.ylabel('MSE')
plt.legend()

# Gráfico de RMSE validación
plt.subplot(1, 3, 2)
plt.plot(val_losses, label='RMSE Validación')
plt.title('RMSE Validación')
plt.xlabel('Épocas')
plt.ylabel('RMSE')
plt.legend()

# Predicciones vs Valores reales
plt.subplot(1, 3, 3)
plt.scatter(y_test_original, y_pred_original, alpha=0.5)
plt.plot([y_test_original.min(), y_test_original.max()], 
         [y_test_original.min(), y_test_original.max()], 
         'r--', lw=2)
plt.xlabel('Valores reales')
plt.ylabel('Predicciones')
plt.title('Predicciones vs Valores reales')
plt.grid(True)

plt.tight_layout()
plt.savefig('california_housing_results.png')
plt.show()

# Mostrar ejemplos de predicciones
print("\nComparativa de predicciones vs valores reales:")
sample_indices = np.random.choice(len(y_test_original), 10, replace=False)

for i, idx in enumerate(sample_indices):
    print(f"Muestra {i+1}: Real = {y_test_original[idx]:.4f}, Predicción = {y_pred_original[idx]:.4f}, " + 
          f"Diff = {abs(y_test_original[idx] - y_pred_original[idx]):.4f}")

# Información GPU
print(f"\nPyTorch versión: {torch.__version__}")
print(f"CUDA disponible: {torch.cuda.is_available()}")
if torch.cuda.is_available():
    print(f"Dispositivo CUDA: {torch.cuda.get_device_name(0)}")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn.preprocessing import StandardScaler, RobustScaler, PowerTransformer
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score
import math
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA
import pandas as pd
import time

# Configuración de semilla para reproducibilidad
torch.manual_seed(42)
np.random.seed(42)

# Verificar disponibilidad de GPU
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Usando: {device}")

# Cargar dataset
print("Cargando y preparando datos...")
dataset = datasets.fetch_california_housing()
print(f"MIN = {np.min(dataset.target)}, MAX = {np.max(dataset.target)}")

# Crear un DataFrame para facilitar la manipulación
X_df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
y = dataset.target

# Verificar las columnas
print("Columnas disponibles en el dataset:")
print(X_df.columns.tolist())
print("Primeras filas del dataset:")
print(X_df.head())

# 1. PREPARACIÓN AVANZADA DE DATOS
# ================================

# Feature Engineering con nombres correctos del dataset California Housing
def create_advanced_features(X):
    """Crea características avanzadas para mejorar el modelo"""
    X_new = X.copy()
    
    # Productos cruzados entre características importantes
    X_new['BEDROOM_PER_ROOM'] = X['AveBedrms'] / (X['AveRooms'] + 1e-6)
    X_new['POPULATION_PER_HOUSEHOLD'] = X['Population'] / (X['AveOccup'] + 1e-6)
    X_new['INCOME_PER_POPULATION'] = X['MedInc'] / (np.log1p(X['Population']) + 1e-6)
    X_new['ROOMS_PER_HOUSEHOLD'] = X['AveRooms'] / (X['AveOccup'] + 1e-6)
    
    # Características no lineales
    X_new['MEDINC_SQ'] = X['MedInc'] ** 2
    X_new['LOG_POPULATION'] = np.log1p(X['Population'])
    X_new['LOG_HOUSEHOLD'] = np.log1p(X['AveOccup'])
    X_new['DISTANCE_TO_OCEAN'] = np.sqrt(X['Longitude']**2 + X['Latitude']**2)
    
    # Location clusters (Distance from major cities)
    # San Francisco: 37.7749° N, 122.4194° W
    X_new['DIST_TO_SF'] = np.sqrt((X['Latitude'] - 37.7749)**2 + (X['Longitude'] - (-122.4194))**2)
    # Los Angeles: 34.0522° N, 118.2437° W
    X_new['DIST_TO_LA'] = np.sqrt((X['Latitude'] - 34.0522)**2 + (X['Longitude'] - (-118.2437))**2)
    # San Diego: 32.7157° N, 117.1611° W
    X_new['DIST_TO_SD'] = np.sqrt((X['Latitude'] - 32.7157)**2 + (X['Longitude'] - (-117.1611))**2)
    
    return X_new

# Aplicar feature engineering directamente
try:
    X_enhanced = create_advanced_features(X_df)
    print(f"Características originales: {X_df.shape[1]}, Características aumentadas: {X_enhanced.shape[1]}")
except Exception as e:
    print(f"Error en feature engineering: {e}")
    # En caso de error, continuamos con las características originales
    X_enhanced = X_df.copy()
    print("Continuando con las características originales debido al error.")

# Detección y manejo de outliers
def handle_outliers(X, y, z_threshold=3.0):
    """Identifica y maneja valores atípicos"""
    # Convertir a array si es dataframe
    if isinstance(X, pd.DataFrame):
        X_array = X.values
    else:
        X_array = X
        
    # Calcular z-score para la variable objetivo
    y_mean, y_std = np.mean(y), np.std(y)
    z_scores = np.abs((y - y_mean) / y_std)
    
    # Identificar no-outliers (dentro del umbral z-score)
    non_outliers = z_scores < z_threshold
    
    print(f"Datos originales: {len(y)}, Después de eliminar outliers: {np.sum(non_outliers)}")
    return X_array[non_outliers], y[non_outliers]

# Aplicar manejo de outliers
X_clean, y_clean = handle_outliers(X_enhanced, y, z_threshold=3.0)

# División de datos
y_quantiles = pd.qcut(y_clean, q=10, labels=False, duplicates='drop')
X_train, X_test, y_train, y_test, q_train, q_test = train_test_split(
    X_clean, y_clean, y_quantiles, test_size=0.2, random_state=42, stratify=y_quantiles
)

# Normalización de datos más robusta
def apply_advanced_scaling(X_train, X_test, y_train, y_test):
    """Aplica técnicas avanzadas de escalado"""
    # Para características - Robust Scaler para manejar outliers restantes
    scaler_X = RobustScaler()
    X_train_scaled = scaler_X.fit_transform(X_train)
    X_test_scaled = scaler_X.transform(X_test)
    
    # Para la variable objetivo - Power Transformer para normalidad
    scaler_y = PowerTransformer(method='yeo-johnson')
    y_train_scaled = scaler_y.fit_transform(y_train.reshape(-1, 1)).flatten()
    y_test_scaled = scaler_y.transform(y_test.reshape(-1, 1)).flatten()
    
    return X_train_scaled, X_test_scaled, y_train_scaled, y_test_scaled, scaler_X, scaler_y

# Aplicar escalado avanzado
X_train_scaled, X_test_scaled, y_train_scaled, y_test_scaled, scaler_X, scaler_y = apply_advanced_scaling(
    X_train, X_test, y_train, y_test
)

# 2. MODELOS AVANZADOS
# ===================

# 2.1 Modelo Deep Neural Network avanzado
class AdvancedHousingNN(nn.Module):
    def __init__(self, input_size, dropout_rate=0.2):
        super(AdvancedHousingNN, self).__init__()
        
        # Capas residuales - inspiradas en ResNet para mejor flujo de gradientes
        self.input_layer = nn.Linear(input_size, 256)
        self.batch_norm1 = nn.BatchNorm1d(256)
        
        # Bloque 1
        self.block1_layer1 = nn.Linear(256, 256)
        self.block1_bn1 = nn.BatchNorm1d(256)
        self.block1_layer2 = nn.Linear(256, 256)
        self.block1_bn2 = nn.BatchNorm1d(256)
        
        # Bloque 2
        self.block2_layer1 = nn.Linear(256, 128)
        self.block2_bn1 = nn.BatchNorm1d(128)
        self.block2_layer2 = nn.Linear(128, 128)
        self.block2_bn2 = nn.BatchNorm1d(128)
        
        # Bloque 3
        self.block3_layer1 = nn.Linear(128, 64)
        self.block3_bn1 = nn.BatchNorm1d(64)
        self.block3_layer2 = nn.Linear(64, 64)
        self.block3_bn2 = nn.BatchNorm1d(64)
        
        # Capa de salida
        self.output_layer = nn.Linear(64, 1)
        
        # Dropout más sofisticado
        self.dropout = nn.Dropout(dropout_rate)
        
        # LeakyReLU para gradientes más estables
        self.activation = nn.LeakyReLU(0.1)
        
        # Inicialización de pesos especializada
        self._initialize_weights()
    
    def _initialize_weights(self):
        for m in self.modules():
            if isinstance(m, nn.Linear):
                # Inicialización He con un pequeño factor de escala para evitar saturación
                nn.init.kaiming_normal_(m.weight, mode='fan_in', nonlinearity='leaky_relu', a=0.1)
                if m.bias is not None:
                    nn.init.zeros_(m.bias)
    
    def forward(self, x):
        # Capa de entrada
        identity = self.activation(self.batch_norm1(self.input_layer(x)))
        x = self.dropout(identity)
        
        # Bloque 1 (residual)
        residual = x
        x = self.activation(self.block1_bn1(self.block1_layer1(x)))
        x = self.block1_bn2(self.block1_layer2(x))
        x = self.activation(x + residual)  # Conexión residual
        x = self.dropout(x)
        
        # Bloque 2
        x = self.activation(self.block2_bn1(self.block2_layer1(x)))
        x = self.activation(self.block2_bn2(self.block2_layer2(x)))
        x = self.dropout(x)
        
        # Bloque 3
        x = self.activation(self.block3_bn1(self.block3_layer1(x)))
        x = self.activation(self.block3_bn2(self.block3_layer2(x)))
        
        # Salida
        return self.output_layer(x).squeeze()

# Convertir a tensores PyTorch
X_train_tensor = torch.FloatTensor(X_train_scaled).to(device)
y_train_tensor = torch.FloatTensor(y_train_scaled).to(device)
X_test_tensor = torch.FloatTensor(X_test_scaled).to(device)
y_test_tensor = torch.FloatTensor(y_test_scaled).to(device)

# Crear DataLoader con tamaño de lote optimizado
batch_size = 32  # Optimizado para rendimiento/memoria
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

# Función de entrenamiento avanzada con técnicas modernas
def train_advanced_model(model, train_loader, val_X, val_y, epochs=300, patience=25):
    print("Iniciando entrenamiento del modelo neural...")
    criterion = nn.HuberLoss(delta=0.5)  # Huber loss es más robusto a outliers que MSE
    optimizer = optim.AdamW(model.parameters(), lr=0.001, weight_decay=1e-4, amsgrad=True)
    
    # Learning rate scheduler con warmup y cosine annealing
    scheduler = optim.lr_scheduler.CosineAnnealingWarmRestarts(
        optimizer, T_0=20, T_mult=2, eta_min=1e-6
    )
    
    train_losses = []
    val_losses = []
    best_val_loss = float('inf')
    counter = 0
    best_model_state = None
    
    # Para OneCycle policy sería una buena alternativa
    # total_steps = epochs * len(train_loader)
    # scheduler = optim.lr_scheduler.OneCycleLR(optimizer, max_lr=0.01, total_steps=total_steps)
    
    for epoch in range(epochs):
        model.train()
        running_loss = 0.0
        
        for inputs, targets in train_loader:
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, targets)
            loss.backward()
            
            # Gradient clipping para estabilidad
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            
            optimizer.step()
            # Si usáramos OneCycle: scheduler.step()
            
            running_loss += loss.item()
        
        # Cosine Annealing step por época
        scheduler.step()
        
        avg_train_loss = running_loss / len(train_loader)
        train_losses.append(avg_train_loss)
        
        # Evaluación en validación
        model.eval()
        with torch.no_grad():
            val_outputs = model(val_X)
            # MSE para evaluación aunque entrenemos con Huber
            val_loss = nn.MSELoss()(val_outputs, val_y)
            val_rmse = torch.sqrt(val_loss).item()
        
        val_losses.append(val_rmse)
        
        # Imprimir progreso cada 20 épocas
        if (epoch + 1) % 20 == 0:
            print(f'Época {epoch+1}/{epochs}, Pérdida: {avg_train_loss:.6f}, Val RMSE: {val_rmse:.6f}')
        
        # Guardar mejor modelo
        if val_rmse < best_val_loss:
            best_val_loss = val_rmse
            best_model_state = model.state_dict().copy()
            counter = 0
        else:
            counter += 1
            if counter >= patience:
                print(f'Early stopping en época {epoch+1}')
                break
    
    # Cargar el mejor modelo
    model.load_state_dict(best_model_state)
    return model, best_val_loss, train_losses, val_losses

# 2.2 Gradient Boosting Regressor para ensamblaje
def train_gbr_model(X_train, y_train):
    print("Entrenando modelo Gradient Boosting...")
    # Preparar pipeline con PCA para reducir dimensionalidad
    pipeline = Pipeline([
        ('poly', PolynomialFeatures(degree=2, include_bias=False, interaction_only=True)),
        ('pca', PCA(n_components=0.95)),  # Mantiene 95% de la varianza
        ('gbr', GradientBoostingRegressor(
            n_estimators=200,
            learning_rate=0.05,
            max_depth=5,
            min_samples_split=5,
            min_samples_leaf=2,
            subsample=0.8,
            random_state=42
        ))
    ])
    
    # Entrenar modelo
    pipeline.fit(X_train, y_train)
    return pipeline

# 3. ENTRENAMIENTO DE MODELOS
# ==========================
# Neural Network
print("\n" + "="*50)
print("ENTRENANDO RED NEURONAL AVANZADA")
print("="*50)
start_time = time.time()

# Instanciar modelo neural
nn_model = AdvancedHousingNN(input_size=X_train_scaled.shape[1], dropout_rate=0.2).to(device)

# Entrenar modelo neural
nn_model, nn_val_loss, train_losses, val_losses = train_advanced_model(
    nn_model, train_loader, X_test_tensor, y_test_tensor, epochs=300
)

nn_time = time.time() - start_time
print(f"Tiempo de entrenamiento NN: {nn_time:.2f} segundos")

# Gradient Boosting Regressor
print("\n" + "="*50)
print("ENTRENANDO GRADIENT BOOSTING REGRESSOR")
print("="*50)
start_time = time.time()

# Entrenar GBR
gbr_model = train_gbr_model(X_train_scaled, y_train_scaled)

# Evaluar GBR
gbr_preds = gbr_model.predict(X_test_scaled)
gbr_mse = mean_squared_error(y_test_scaled, gbr_preds)
gbr_rmse = np.sqrt(gbr_mse)
print(f"GBR RMSE (escalado): {gbr_rmse:.6f}")

gbr_time = time.time() - start_time
print(f"Tiempo de entrenamiento GBR: {gbr_time:.2f} segundos")

# 4. ENSAMBLAJE DE MODELOS
# =======================
print("\n" + "="*50)
print("CREANDO ENSAMBLAJE DE MODELOS")
print("="*50)

# Generar predicciones de ambos modelos
nn_model.eval()
with torch.no_grad():
    nn_preds = nn_model(X_test_tensor).cpu().numpy()

# Optimizar pesos de ensamblaje
def optimize_ensemble_weights(y_true, pred1, pred2):
    """Encuentra el peso óptimo para combinar dos modelos"""
    best_rmse = float('inf')
    best_weight = 0.5
    
    for weight in np.linspace(0, 1, 101):  # 0.00, 0.01, 0.02, ..., 1.00
        ensemble_pred = weight * pred1 + (1 - weight) * pred2
        rmse = np.sqrt(mean_squared_error(y_true, ensemble_pred))
        
        if rmse < best_rmse:
            best_rmse = rmse
            best_weight = weight
            
    return best_weight, best_rmse

# Encontrar el mejor peso para el ensamblaje
best_weight, ensemble_rmse_scaled = optimize_ensemble_weights(y_test_scaled, nn_preds, gbr_preds)
print(f"Mejor peso para NN: {best_weight:.4f}, GBR: {1-best_weight:.4f}")
print(f"Ensemble RMSE (escalado): {ensemble_rmse_scaled:.6f}")

# Crear predicciones del ensamblaje
ensemble_preds_scaled = best_weight * nn_preds + (1 - best_weight) * gbr_preds

# 5. EVALUACIÓN DE RESULTADOS
# =========================
print("\n" + "="*50)
print("EVALUACIÓN FINAL DE MODELOS")
print("="*50)

# Función para calcular todas las métricas
def calculate_metrics(y_true_scaled, y_pred_scaled, scaler):
    """Calcula múltiples métricas tanto en escala original como escalada"""
    # Métricas escaladas
    mse_scaled = mean_squared_error(y_true_scaled, y_pred_scaled)
    rmse_scaled = np.sqrt(mse_scaled)
    r2_scaled = r2_score(y_true_scaled, y_pred_scaled)
    
    # Convertir de vuelta a la escala original
    y_true_orig = scaler.inverse_transform(y_true_scaled.reshape(-1, 1)).flatten()
    y_pred_orig = scaler.inverse_transform(y_pred_scaled.reshape(-1, 1)).flatten()
    
    # Métricas originales
    mse_orig = mean_squared_error(y_true_orig, y_pred_orig)
    rmse_orig = np.sqrt(mse_orig)
    r2_orig = r2_score(y_true_orig, y_pred_orig)
    
    # Error porcentual absoluto medio (MAPE)
    mape = np.mean(np.abs((y_true_orig - y_pred_orig) / (y_true_orig + 1e-6))) * 100
    
    return {
        'scaled': {'mse': mse_scaled, 'rmse': rmse_scaled, 'r2': r2_scaled},
        'original': {'mse': mse_orig, 'rmse': rmse_orig, 'r2': r2_orig, 'mape': mape},
        'predictions': {'true': y_true_orig, 'pred': y_pred_orig}
    }

# Evaluar modelos individuales
nn_metrics = calculate_metrics(y_test_scaled, nn_preds, scaler_y)
gbr_metrics = calculate_metrics(y_test_scaled, gbr_preds, scaler_y)
ensemble_metrics = calculate_metrics(y_test_scaled, ensemble_preds_scaled, scaler_y)

# Imprimir resultados
models = ['Neural Network', 'Gradient Boosting', 'Ensemble']
all_metrics = [nn_metrics, gbr_metrics, ensemble_metrics]

for name, metrics in zip(models, all_metrics):
    print(f"\n{name}:")
    print(f"  RMSE (escalado): {metrics['scaled']['rmse']:.6f}")
    print(f"  RMSE (original): {metrics['original']['rmse']:.6f}")
    print(f"  R² (original): {metrics['original']['r2']:.6f}")
    print(f"  MAPE: {metrics['original']['mape']:.2f}%")

# 6. VISUALIZACIONES
# ================
plt.figure(figsize=(18, 12))

# Gráfico 1: Pérdida de entrenamiento (Neural Network)
plt.subplot(2, 3, 1)
plt.plot(train_losses)
plt.title('Pérdida de Entrenamiento (NN)')
plt.xlabel('Épocas')
plt.ylabel('Pérdida')

# Gráfico 2: RMSE de validación (Neural Network)
plt.subplot(2, 3, 2)
plt.plot(val_losses)
plt.title('RMSE de Validación (NN)')
plt.xlabel('Épocas')
plt.ylabel('RMSE')

# Gráfico 3: Comparativa de predicciones (Neural Network)
plt.subplot(2, 3, 3)
plt.scatter(nn_metrics['predictions']['true'], nn_metrics['predictions']['pred'], alpha=0.5)
plt.plot([min(nn_metrics['predictions']['true']), max(nn_metrics['predictions']['true'])],
         [min(nn_metrics['predictions']['true']), max(nn_metrics['predictions']['true'])], 'r--')
plt.title('Predicciones vs Real (NN)')
plt.xlabel('Valores Reales')
plt.ylabel('Predicciones')

# Gráfico 4: Comparativa de predicciones (GBR)
plt.subplot(2, 3, 4)
plt.scatter(gbr_metrics['predictions']['true'], gbr_metrics['predictions']['pred'], alpha=0.5)
plt.plot([min(gbr_metrics['predictions']['true']), max(gbr_metrics['predictions']['true'])],
         [min(gbr_metrics['predictions']['true']), max(gbr_metrics['predictions']['true'])], 'r--')
plt.title('Predicciones vs Real (GBR)')
plt.xlabel('Valores Reales')
plt.ylabel('Predicciones')

# Gráfico 5: Comparativa de predicciones (Ensemble)
plt.subplot(2, 3, 5)
plt.scatter(ensemble_metrics['predictions']['true'], ensemble_metrics['predictions']['pred'], alpha=0.5)
plt.plot([min(ensemble_metrics['predictions']['true']), max(ensemble_metrics['predictions']['true'])],
         [min(ensemble_metrics['predictions']['true']), max(ensemble_metrics['predictions']['true'])], 'r--')
plt.title('Predicciones vs Real (Ensemble)')
plt.xlabel('Valores Reales')
plt.ylabel('Predicciones')

# Gráfico 6: Comparativa de errores
plt.subplot(2, 3, 6)
models_names = ['NN', 'GBR', 'Ensemble']
original_rmse = [m['original']['rmse'] for m in all_metrics]

plt.bar(models_names, original_rmse)
plt.title('RMSE por Modelo')
plt.ylabel('RMSE (no escalado)')

plt.tight_layout()
plt.savefig('california_housing_advanced_results.png')
plt.show()

# 7. GUARDAR MODELOS E INFORMACIÓN PARA PREDICCIÓN
# ==============================================
import pickle

# Guardar modelo neuronal
torch.save({
    'model_state_dict': nn_model.state_dict(),
    'input_size': X_train_scaled.shape[1],
    'dropout_rate': 0.2
}, 'california_nn_model.pth')

# Guardar GBR y scaler
with open('california_gbr_model.pkl', 'wb') as f:
    pickle.dump({
        'gbr_model': gbr_model,
        'scaler_X': scaler_X,
        'scaler_y': scaler_y,
        'ensemble_weight': best_weight
    }, f)

# Función para predicciones con el ensamblaje
def predict_housing_price(features, nn_model, gbr_model, scaler_X, scaler_y, ensemble_weight=0.5):
    """
    Predice precio de vivienda usando el modelo ensamblado
    
    Args:
        features: Características sin procesar
        nn_model: Modelo de red neuronal entrenado
        gbr_model: Modelo GBR entrenado
        scaler_X: Scaler para normalizar características
        scaler_y: Scaler para desnormalizar predicciones
        ensemble_weight: Peso óptimo para el ensamblaje
    
    Returns:
        float: Precio de vivienda predicho
    """
    # Preparar características
    if isinstance(features, pd.DataFrame):
        # Aplicar feature engineering si es necesario
        if features.shape[1] == 8:  # Si son solo las características originales
            features = create_advanced_features(features)
    
    # Normalizar características
    features_scaled = scaler_X.transform(features.reshape(1, -1) if features.ndim == 1 else features)
    features_tensor = torch.FloatTensor(features_scaled).to(device)
    
    # Predicción con NN
    nn_model.eval()
    with torch.no_grad():
        nn_pred_scaled = nn_model(features_tensor).cpu().numpy()
    
    # Predicción con GBR
    gbr_pred_scaled = gbr_model.predict(features_scaled)
    
    # Ensamblaje
    ensemble_pred_scaled = ensemble_weight * nn_pred_scaled + (1 - ensemble_weight) * gbr_pred_scaled
    
    # Desnormalizar
    ensemble_pred = scaler_y.inverse_transform(
        ensemble_pred_scaled.reshape(-1, 1)
    ).flatten()
    
    return ensemble_pred[0]

# Mostrar información GPU
print(f"\nPyTorch versión: {torch.__version__}")
print(f"CUDA disponible: {torch.cuda.is_available()}")
if torch.cuda.is_available():
    print(f"Dispositivo CUDA: {torch.cuda.get_device_name(0)}")
    print(f"Memoria GPU utilizada: {torch.cuda.memory_allocated(0) / 1024**2:.2f} MB")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import numpy as np
from sklearn.model_selection import train_test_split, KFold
from sklearn import datasets
from sklearn.preprocessing import StandardScaler, RobustScaler, PowerTransformer
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import math
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor, ExtraTreesRegressor
from xgboost import XGBRegressor
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA
import pandas as pd
import time
import pickle
from scipy.stats import probplot

# Configuración de semilla para reproducibilidad
torch.manual_seed(42)
np.random.seed(42)

# Verificar disponibilidad de GPU
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Usando: {device}")

# Cargar dataset
print("Cargando y preparando datos...")
dataset = datasets.fetch_california_housing()
print(f"MIN = {np.min(dataset.target)}, MAX = {np.max(dataset.target)}")

# Crear un DataFrame para facilitar la manipulación
X_df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
y = dataset.target

# Verificar las columnas
print("Columnas disponibles en el dataset:")
print(X_df.columns.tolist())
print("Primeras filas del dataset:")
print(X_df.head())

# 1. PREPARACIÓN AVANZADA DE DATOS
# ================================

# Feature Engineering con transformaciones mejoradas
def create_enhanced_features(X):
    """Crea características mejoradas con transformaciones más específicas"""
    X_new = X.copy()
    
    # Características de relación
    X_new['BEDROOM_PER_ROOM'] = X['AveBedrms'] / (X['AveRooms'] + 1e-6)
    X_new['POPULATION_PER_HOUSEHOLD'] = X['Population'] / (X['AveOccup'] + 1e-6)
    X_new['ROOMS_PER_HOUSEHOLD'] = X['AveRooms'] / (X['AveOccup'] + 1e-6)
    
    # Características de ingresos
    X_new['MEDINC_SQ'] = X['MedInc'] ** 2
    X_new['MEDINC_SQRT'] = np.sqrt(X['MedInc'] + 1e-6)
    X_new['INCOME_PER_POPULATION'] = X['MedInc'] / (np.log1p(X['Population']) + 1e-6)
    X_new['INCOME_PER_HOUSEHOLD'] = X['MedInc'] / (X['AveOccup'] + 1e-6)
    X_new['INCOME_PER_ROOM'] = X['MedInc'] / (X['AveRooms'] + 1e-6)
    
    # Transformaciones logarítmicas
    X_new['LOG_MEDINC'] = np.log1p(X['MedInc'])
    X_new['LOG_POPULATION'] = np.log1p(X['Population'])
    X_new['LOG_HOUSEHOLD'] = np.log1p(X['AveOccup'])
    X_new['LOG_ROOMS'] = np.log1p(X['AveRooms'])
    
    # Características de distancia y geolocalización
    X_new['DISTANCE_TO_OCEAN'] = np.sqrt(X['Longitude']**2 + X['Latitude']**2)
    
    # Características de proximidad con ponderación exponencial (kernel gaussiano)
    # San Francisco: 37.7749° N, 122.4194° W
    sf_dist = np.sqrt((X['Latitude'] - 37.7749)**2 + (X['Longitude'] - (-122.4194))**2)
    X_new['DIST_TO_SF'] = sf_dist
    X_new['DIST_TO_SF_EXP'] = np.exp(-0.5 * sf_dist)
    
    # Los Angeles: 34.0522° N, 118.2437° W
    la_dist = np.sqrt((X['Latitude'] - 34.0522)**2 + (X['Longitude'] - (-118.2437))**2)
    X_new['DIST_TO_LA'] = la_dist
    X_new['DIST_TO_LA_EXP'] = np.exp(-0.5 * la_dist)
    
    # San Diego: 32.7157° N, 117.1611° W
    sd_dist = np.sqrt((X['Latitude'] - 32.7157)**2 + (X['Longitude'] - (-117.1611))**2)
    X_new['DIST_TO_SD'] = sd_dist
    X_new['DIST_TO_SD_EXP'] = np.exp(-0.5 * sd_dist)
    
    # Interacciones importantes
    X_new['MEDINC_X_ROOMS'] = X['MedInc'] * X['AveRooms']
    X_new['MEDINC_X_HOUSEAGE'] = X['MedInc'] * X['HouseAge']
    X_new['AGE_X_ROOMS'] = X['HouseAge'] * X['AveRooms']
    
    # Características de "proximidad a la costa"
    # La costa oeste de California está aproximadamente en longitud -124
    X_new['COASTAL_PROXIMITY'] = -(X['Longitude'] + 124)  # Más positivo = más cerca de la costa
    X_new['IS_COASTAL'] = (X['Longitude'] > -122).astype(float)  # Indicador binario de proximidad a costa
    
    # Combinaciones no lineales
    X_new['ROOMS_X_COASTAL'] = X['AveRooms'] * X_new['IS_COASTAL']
    X_new['INCOME_X_COASTAL'] = X['MedInc'] * X_new['IS_COASTAL']
    
    return X_new

# Aplicar feature engineering mejorado
try:
    X_enhanced = create_enhanced_features(X_df)
    print(f"Características originales: {X_df.shape[1]}, Características aumentadas: {X_enhanced.shape[1]}")
except Exception as e:
    print(f"Error en feature engineering: {e}")
    X_enhanced = X_df.copy()
    print("Continuando con las características originales debido al error.")

# Mejor detección y manejo de outliers que considera relaciones multivariables
def handle_outliers_advanced(X, y, z_threshold=3.0, isolation_threshold=0.02):
    """Identifica y maneja valores atípicos con método combinado"""
    from sklearn.ensemble import IsolationForest
    
    # Convertir a array si es dataframe
    if isinstance(X, pd.DataFrame):
        X_array = X.values
    else:
        X_array = X
    
    # 1. Z-score para la variable objetivo
    y_mean, y_std = np.mean(y), np.std(y)
    z_scores = np.abs((y - y_mean) / y_std)
    non_outliers_z = z_scores < z_threshold
    
    # 2. Isolation Forest para detección multivariable
    iso_forest = IsolationForest(contamination=isolation_threshold, random_state=42)
    outlier_pred = iso_forest.fit_predict(X_array)
    non_outliers_iso = outlier_pred == 1  # 1 = inlier, -1 = outlier
    
    # Combinar ambos métodos (conservador - mantiene punto si uno de los métodos lo considera válido)
    combined_mask = non_outliers_z | non_outliers_iso
    
    print(f"Datos originales: {len(y)}")
    print(f"Outliers por Z-score: {len(y) - np.sum(non_outliers_z)}")
    print(f"Outliers por Isolation Forest: {len(y) - np.sum(non_outliers_iso)}")
    print(f"Después de eliminar outliers: {np.sum(combined_mask)}")
    
    return X_array[combined_mask], y[combined_mask]

# Aplicar manejo de outliers avanzado
X_clean, y_clean = handle_outliers_advanced(X_enhanced, y, z_threshold=3.0, isolation_threshold=0.03)

# División de datos estratificada mejorada
def enhanced_stratified_split(X, y, test_size=0.2):
    """Divide los datos con estratificación mejorada por valor y ubicación"""
    
    # 1. Estratificación por valor (cuantiles)
    y_quantiles = pd.qcut(y, q=15, labels=False, duplicates='drop')
    
    # 2. Estratificación por ubicación (norte/sur/centro de California)
    latitude = X[:, X_df.columns.get_loc('Latitude')]
    location_strata = pd.cut(latitude, bins=[30, 35, 38, 42], labels=[0, 1, 2], include_lowest=True)
    
    # 3. Combinación de ambos estratos
    combined_strata = pd.Series([f"{q}_{loc}" for q, loc in zip(y_quantiles, location_strata)])
    
    # Realizar split estratificado
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=42, stratify=combined_strata
    )
    
    print(f"Train set: {X_train.shape[0]} muestras, Test set: {X_test.shape[0]} muestras")
    
    return X_train, X_test, y_train, y_test

# Aplicar división estratificada mejorada
X_train, X_test, y_train, y_test = enhanced_stratified_split(X_clean, y_clean, test_size=0.2)

# Técnicas de aumento de datos para regresión
def augment_regression_data(X, y, noise_level=0.02, n_samples=1000):
    """Aumenta los datos de regresión con ejemplos sintéticos"""
    
    # Seleccionar índices aleatorios
    indices = np.random.choice(len(X), size=n_samples, replace=True)
    
    # Crear copias
    X_aug = X[indices].copy()
    y_aug = y[indices].copy()
    
    # Añadir ruido gaussiano a las características
    noise = np.random.normal(0, noise_level, X_aug.shape)
    X_aug += noise
    
    # Ajustar ligeramente los valores objetivo basados en el ruido añadido
    y_noise = np.random.normal(0, noise_level/2, y_aug.shape)
    y_aug += y_noise
    
    # Combinar con datos originales
    X_combined = np.vstack([X, X_aug])
    y_combined = np.concatenate([y, y_aug])
    
    print(f"Datos aumentados: original {len(X)} + sintético {n_samples} = {len(X_combined)}")
    
    return X_combined, y_combined

# Aplicar aumento de datos (descomentarlo si deseas usarlo)
# X_train, y_train = augment_regression_data(X_train, y_train, noise_level=0.01, n_samples=int(len(X_train)*0.3))

# Escalado avanzado con transformación de características
def apply_advanced_scaling(X_train, X_test, y_train, y_test):
    """Aplica técnicas avanzadas de escalado con transformación para normalidad"""
    
    # Para características - Robust Scaler para manejar outliers restantes
    scaler_X = RobustScaler()
    X_train_scaled = scaler_X.fit_transform(X_train)
    X_test_scaled = scaler_X.transform(X_test)
    
    # Para la variable objetivo - Power Transformer para normalidad
    scaler_y = PowerTransformer(method='yeo-johnson')
    y_train_scaled = scaler_y.fit_transform(y_train.reshape(-1, 1)).flatten()
    y_test_scaled = scaler_y.transform(y_test.reshape(-1, 1)).flatten()
    
    return X_train_scaled, X_test_scaled, y_train_scaled, y_test_scaled, scaler_X, scaler_y

# Aplicar escalado avanzado
X_train_scaled, X_test_scaled, y_train_scaled, y_test_scaled, scaler_X, scaler_y = apply_advanced_scaling(
    X_train, X_test, y_train, y_test
)

# 2. MODELOS AVANZADOS
# ===================

# 2.1 Modelo Deep Neural Network mejorado con arquitectura avanzada
class EnhancedHousingNN(nn.Module):
    def __init__(self, input_size, initial_units=320, dropout_rate=0.15, depth_factor=0.5):
        super(EnhancedHousingNN, self).__init__()
        
        # Aumentar capa inicial para capturar más patrones
        self.input_layer = nn.Linear(input_size, initial_units)
        self.batch_norm1 = nn.BatchNorm1d(initial_units)
        
        # Bloques residuales más grandes con conexiones skip
        self.block1 = nn.Sequential(
            nn.Linear(initial_units, initial_units),
            nn.BatchNorm1d(initial_units),
            nn.LeakyReLU(0.1),
            nn.Dropout(dropout_rate),
            nn.Linear(initial_units, initial_units),
            nn.BatchNorm1d(initial_units)
        )
        
        # Segunda capa con menor dimensionalidad
        second_layer_size = int(initial_units * depth_factor)
        self.block2 = nn.Sequential(
            nn.Linear(initial_units, second_layer_size),
            nn.BatchNorm1d(second_layer_size),
            nn.LeakyReLU(0.1),
            nn.Dropout(dropout_rate),
            nn.Linear(second_layer_size, second_layer_size),
            nn.BatchNorm1d(second_layer_size)
        )
        
        # Tercera capa con menor dimensionalidad aún
        third_layer_size = int(second_layer_size * depth_factor)
        self.final_block = nn.Sequential(
            nn.Linear(second_layer_size, third_layer_size),
            nn.BatchNorm1d(third_layer_size),
            nn.LeakyReLU(0.1),
            nn.Linear(third_layer_size, third_layer_size // 2),
            nn.BatchNorm1d(third_layer_size // 2),
            nn.LeakyReLU(0.1)
        )
        
        self.output_layer = nn.Linear(third_layer_size // 2, 1)
        self.dropout = nn.Dropout(dropout_rate)
        self.activation = nn.LeakyReLU(0.1)
        
        self._initialize_weights()
        
    def _initialize_weights(self):
        # Inicialización con Kaiming pero con un factor de escala para mayor estabilidad
        for m in self.modules():
            if isinstance(m, nn.Linear):
                nn.init.kaiming_normal_(m.weight, mode='fan_out', nonlinearity='leaky_relu', a=0.1)
                if m.bias is not None:
                    nn.init.constant_(m.bias, 0.01)  # Pequeño valor positivo en lugar de cero
    
    def forward(self, x):
        # Capa inicial
        x = self.activation(self.batch_norm1(self.input_layer(x)))
        x = self.dropout(x)
        
        # Bloque residual 1
        identity = x
        x_block1 = self.block1(x)
        x = self.activation(x_block1 + identity)  # Conexión residual
        
        # Bloque 2
        x = self.block2(x)
        
        # Capas finales
        x = self.final_block(x)
        
        # Salida
        return self.output_layer(x).squeeze()

# Convertir a tensores PyTorch
X_train_tensor = torch.FloatTensor(X_train_scaled).to(device)
y_train_tensor = torch.FloatTensor(y_train_scaled).to(device)
X_test_tensor = torch.FloatTensor(X_test_scaled).to(device)
y_test_tensor = torch.FloatTensor(y_test_scaled).to(device)

# Crear DataLoader con tamaño de lote optimizado
batch_size = 64  # Valor recomendado, ajustar según hardware
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

# Función de entrenamiento avanzada con técnicas modernas
def train_advanced_model_v2(model, train_loader, val_X, val_y, epochs=350, patience=30):
    print("Iniciando entrenamiento del modelo neuronal mejorado...")
    
    # Huber loss con parámetro delta más bajo para mayor robustez con outliers
    criterion = nn.HuberLoss(delta=0.3)
    
    # Optimizer con parámetros optimizados
    optimizer = optim.AdamW(
        model.parameters(), 
        lr=0.0008,  # Menor learning rate inicial
        weight_decay=2e-5,  # Más regularización
        betas=(0.9, 0.999),
        amsgrad=True
    )
    
    # Learning rate scheduler con warmup y cosine annealing
    def lr_lambda(current_step):
        warmup_steps = 5 * len(train_loader)  # 5 épocas de warmup
        if current_step < warmup_steps:
            return float(current_step) / float(max(1, warmup_steps))
        return 0.5 * (1.0 + np.cos((current_step - warmup_steps) / 
                                   (epochs * len(train_loader) - warmup_steps) * np.pi))
    
    scheduler = optim.lr_scheduler.LambdaLR(optimizer, lr_lambda)
    
    train_losses = []
    val_losses = []
    best_val_loss = float('inf')
    counter = 0
    best_model_state = None
    
    # Implementar SWA (Stochastic Weight Averaging) para mejor generalización
    swa_model = torch.optim.swa_utils.AveragedModel(model)
    swa_start = int(epochs * 0.75)  # Comenzar SWA en el último 25% del entrenamiento
    swa_scheduler = torch.optim.swa_utils.SWALR(
        optimizer, swa_lr=0.0005, anneal_strategy="cos", anneal_epochs=10
    )
    
    for epoch in range(epochs):
        model.train()
        running_loss = 0.0
        
        for inputs, targets in train_loader:
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, targets)
            
            # Regularización L1 adicional
            l1_lambda = 1e-5
            l1_norm = sum(p.abs().sum() for p in model.parameters())
            loss = loss + l1_lambda * l1_norm
            
            loss.backward()
            
            # Gradient clipping
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=0.8)
            
            optimizer.step()
            scheduler.step()
            
            running_loss += loss.item()
        
        # Aplicar SWA en las últimas épocas
        if epoch >= swa_start:
            swa_model.update_parameters(model)
            swa_scheduler.step()
        else:
            scheduler.step()
        
        avg_train_loss = running_loss / len(train_loader)
        train_losses.append(avg_train_loss)
        
        # Evaluación en validación
        model.eval()
        with torch.no_grad():
            val_outputs = model(val_X)
            val_loss = nn.MSELoss()(val_outputs, val_y)
            val_rmse = torch.sqrt(val_loss).item()
        
        val_losses.append(val_rmse)
        
        if (epoch + 1) % 20 == 0:
            print(f'Época {epoch+1}/{epochs}, Pérdida: {avg_train_loss:.6f}, Val RMSE: {val_rmse:.6f}')
        
        if val_rmse < best_val_loss:
            best_val_loss = val_rmse
            best_model_state = model.state_dict().copy()
            counter = 0
        else:
            counter += 1
            if counter >= patience:
                print(f'Early stopping en época {epoch+1}')
                break
    
    # Finalizar SWA y actualizar BN
    if epoch >= swa_start:
        # Crear dataloader para actualizar BatchNorm
        update_loader = DataLoader(TensorDataset(X_train_tensor[:1000], y_train_tensor[:1000]), 
                                 batch_size=batch_size)
        torch.optim.swa_utils.update_bn(update_loader, swa_model)
        swa_model_state = swa_model.module.state_dict()
        
        # Comprobar si SWA es mejor
        model.load_state_dict(swa_model_state)
        model.eval()
        with torch.no_grad():
            swa_val_outputs = model(val_X)
            swa_val_loss = nn.MSELoss()(swa_val_outputs, val_y)
            swa_val_rmse = torch.sqrt(swa_val_loss).item()
        
        if swa_val_rmse < best_val_loss:
            print(f"SWA mejora el rendimiento: {swa_val_rmse:.6f} vs {best_val_loss:.6f}")
            best_model_state = swa_model_state
            best_val_loss = swa_val_rmse
    
    # Cargar el mejor modelo
    model.load_state_dict(best_model_state)
    return model, best_val_loss, train_losses, val_losses

# Función para entrenar múltiples modelos y crear un ensamblaje
def train_multi_model_ensemble(X_train, y_train, X_test, y_test, nn_model):
    print("Entrenando ensamblaje con múltiples modelos...")
    
    # 1. Gradient Boosting (con características polinómicas y PCA)
    print("Entrenando Gradient Boosting...")
    gbr_pipeline = Pipeline([
        ('poly', PolynomialFeatures(degree=2, include_bias=False, interaction_only=True)),
        ('pca', PCA(n_components=0.95)),
        ('gbr', GradientBoostingRegressor(
            n_estimators=200,
            learning_rate=0.05,
            max_depth=5,
            min_samples_split=5,
            min_samples_leaf=2,
            subsample=0.8,
            random_state=42
        ))
    ])
    gbr_pipeline.fit(X_train, y_train)
    gbr_preds = gbr_pipeline.predict(X_test)
    
    # 2. Random Forest
    print("Entrenando Random Forest...")
    rf_model = RandomForestRegressor(
        n_estimators=200, 
        max_depth=12,
        min_samples_split=5,
        min_samples_leaf=2, 
        n_jobs=-1,
        random_state=42
    )
    rf_model.fit(X_train, y_train)
    rf_preds = rf_model.predict(X_test)
    
    # 3. Extra Trees
    print("Entrenando Extra Trees...")
    et_model = ExtraTreesRegressor(
        n_estimators=200,
        max_depth=12,
        min_samples_split=4,
        min_samples_leaf=2,
        n_jobs=-1,
        random_state=42
    )
    et_model.fit(X_train, y_train)
    et_preds = et_model.predict(X_test)
    
    # 4. XGBoost
    print("Entrenando XGBoost...")
    xgb_model = XGBRegressor(
        n_estimators=200,
        max_depth=6,
        learning_rate=0.05,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=42
    )
    xgb_model.fit(X_train, y_train)
    xgb_preds = xgb_model.predict(X_test)
    
    # 5. Neural Network (ya entrenado, pasado como parámetro)
    print("Generando predicciones con el modelo neural...")
    nn_model.eval()
    with torch.no_grad():
        nn_preds = nn_model(torch.FloatTensor(X_test).to(device)).cpu().numpy()
    
    # Optimización de ensamblaje con ajuste de pesos avanzado
    def optimize_ensemble_weights_v2(y_true, predictions_list, model_names):
        from scipy.optimize import minimize
        
        # Función para minimizar
        def rmse_loss(weights):
            # Normalizar pesos para que sumen 1
            weights = weights / np.sum(weights)
            
            # Calcular predicción ensamblada
            ensemble_pred = np.zeros_like(y_true)
            for i, preds in enumerate(predictions_list):
                ensemble_pred += weights[i] * preds
            
            # Calcular RMSE
            return np.sqrt(mean_squared_error(y_true, ensemble_pred))
        
        # Restricciones: los pesos deben ser positivos
        constraints = ({'type': 'ineq', 'fun': lambda w: w})
        
        # Inicializar con pesos iguales
        initial_weights = np.ones(len(predictions_list)) / len(predictions_list)
        
        # Optimizar
        result = minimize(
            rmse_loss, 
            initial_weights, 
            method='SLSQP',
            constraints=constraints,
            bounds=[(0.0, 1.0) for _ in range(len(predictions_list))]
        )
        
        # Normalizar pesos finales
        best_weights = result.x / np.sum(result.x)
        best_rmse = result.fun
        
        # Mostrar pesos
        print("Pesos optimizados para el ensamblaje:")
        for name, weight in zip(model_names, best_weights):
            print(f"  {name}: {weight:.4f}")
        
        return best_weights, best_rmse
    
    # Lista de predicciones y nombres de modelos
    predictions = [nn_preds, gbr_preds, rf_preds, et_preds, xgb_preds]
    model_names = ["Neural Network", "Gradient Boosting", "Random Forest", 
                  "Extra Trees", "XGBoost"]
    
    # Encontrar pesos óptimos
    best_weights, ensemble_rmse = optimize_ensemble_weights_v2(
        y_test, predictions, model_names
    )
    print(f"Ensemble RMSE: {ensemble_rmse:.6f}")
    
    # Crear predicción final ensamblada
    ensemble_pred = np.zeros_like(y_test)
    for i, preds in enumerate(predictions):
        ensemble_pred += best_weights[i] * preds
    
    return {
        'ensemble_pred': ensemble_pred,
        'ensemble_rmse': ensemble_rmse,
        'models': {
            'gbr': gbr_pipeline,
            'rf': rf_model,
            'et': et_model,
            'xgb': xgb_model
        },
        'weights': dict(zip(model_names, best_weights)),
        'predictions': {
            'nn': nn_preds,
            'gbr': gbr_preds,
            'rf': rf_preds,
            'et': et_preds,
            'xgb': xgb_preds
        }
    }

# Función para evaluar y visualizar resultados finales
def evaluate_and_visualize_final_model(ensemble_results, y_test, scaler_y, nn_train_losses, nn_val_losses):
    """Evaluación detallada y visualizaciones avanzadas del modelo final"""
    
    # Extraer predicciones
    ensemble_pred = ensemble_results['ensemble_pred']
    nn_preds = ensemble_results['predictions']['nn']
    gbr_preds = ensemble_results['predictions']['gbr']
    rf_preds = ensemble_results['predictions']['rf']
    et_preds = ensemble_results['predictions']['et']
    xgb_preds = ensemble_results['predictions']['xgb']
    
    # Transformar a escala original
    y_test_orig = scaler_y.inverse_transform(y_test.reshape(-1, 1)).flatten()
    ensemble_pred_orig = scaler_y.inverse_transform(ensemble_pred.reshape(-1, 1)).flatten()
    nn_preds_orig = scaler_y.inverse_transform(nn_preds.reshape(-1, 1)).flatten()
    gbr_preds_orig = scaler_y.inverse_transform(gbr_preds.reshape(-1, 1)).flatten()
    
    # Calcular métricas
    ensemble_mse = mean_squared_error(y_test_orig, ensemble_pred_orig)
    ensemble_rmse = np.sqrt(ensemble_mse)
    ensemble_r2 = r2_score(y_test_orig, ensemble_pred_orig)
    ensemble_mae = mean_absolute_error(y_test_orig, ensemble_pred_orig)
    
    # MAPE (Mean Absolute Percentage Error)
    ensemble_mape = np.mean(np.abs((y_test_orig - ensemble_pred_orig) / (y_test_orig + 1e-6))) * 100
    
    # Métricas por modelo individual
    nn_rmse = np.sqrt(mean_squared_error(y_test_orig, nn_preds_orig))
    gbr_rmse = np.sqrt(mean_squared_error(y_test_orig, gbr_preds_orig))
    
    print("\nMétricas finales del ensamblaje:")
    print(f"  RMSE: {ensemble_rmse:.4f}")
    print(f"  MSE: {ensemble_mse:.4f}")
    print(f"  MAE: {ensemble_mae:.4f}")
    print(f"  R²: {ensemble_r2:.4f}")
    print(f"  MAPE: {ensemble_mape:.2f}%")
    
    # Comparar con modelos individuales
    print("\nRMSE por modelo:")
    print(f"  Neural Network: {nn_rmse:.4f}")
    print(f"  Gradient Boosting: {gbr_rmse:.4f}")
    print(f"  Ensemble: {ensemble_rmse:.4f}")
    
    # Visualizaciones avanzadas
    plt.figure(figsize=(20, 15))
    
    # 1. Curva de aprendizaje y validación
    plt.subplot(3, 2, 1)
    plt.plot(nn_train_losses, label='Pérdida entrenamiento')
    plt.title('Curva de aprendizaje (NN)')
    plt.xlabel('Épocas')
    plt.ylabel('Pérdida')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    plt.subplot(3, 2, 2)
    plt.plot(nn_val_losses, label='RMSE validación')
    plt.title('Curva de validación (NN)')
    plt.xlabel('Épocas')
    plt.ylabel('RMSE')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    # 2. Predicciones vs Real con densidad
    plt.subplot(3, 2, 3)
    h = plt.hist2d(y_test_orig, ensemble_pred_orig, bins=50, cmap='viridis')
    plt.colorbar(h[3], label='Densidad')
    
    # Línea de referencia
    min_val, max_val = min(y_test_orig.min(), ensemble_pred_orig.min()), max(y_test_orig.max(), ensemble_pred_orig.max())
    plt.plot([min_val, max_val], [min_val, max_val], 'r--', label='Ideal')
    plt.title('Predicciones vs Valores Reales (Ensemble)')
    plt.xlabel('Valores Reales')
    plt.ylabel('Predicciones')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    # 3. Distribución de errores
    plt.subplot(3, 2, 4)
    errors = ensemble_pred_orig - y_test_orig
    plt.hist(errors, bins=50, alpha=0.7, color='blue')
    plt.axvline(x=0, color='r', linestyle='--')
    plt.title('Distribución de Errores')
    plt.xlabel('Error (Predicción - Real)')
    plt.ylabel('Frecuencia')
    plt.grid(True, alpha=0.3)
    
    # 4. Error vs Valor Real
    plt.subplot(3, 2, 5)
    plt.scatter(y_test_orig, np.abs(errors), alpha=0.5, s=10)
    plt.title('Magnitud de Error vs Valor Real')
    plt.xlabel('Valor Real')
    plt.ylabel('Error Absoluto')
    plt.grid(True, alpha=0.3)
    
    # 5. Gráfico QQ para normalidad de errores
    plt.subplot(3, 2, 6)
    probplot(errors, dist="norm", plot=plt)
    plt.title('Gráfico Q-Q de Errores')
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig('advanced_error_analysis.png', dpi=300)
    plt.show()
    
    # Comparación de todos los modelos
    plt.figure(figsize=(16, 8))
    
    # Panel 1: Comparar predicciones de todos los modelos
    plt.subplot(1, 2, 1)
    models_names = ['NN', 'GBR', 'RF', 'ET', 'XGB', 'Ensemble']
    all_preds = [nn_preds, gbr_preds, rf_preds, et_preds, xgb_preds, ensemble_pred]
    all_rmses = []
    
    for preds in all_preds:
        preds_orig = scaler_y.inverse_transform(preds.reshape(-1, 1)).flatten()
        rmse = np.sqrt(mean_squared_error(y_test_orig, preds_orig))
        all_rmses.append(rmse)
    
    plt.bar(models_names, all_rmses)
    plt.title('RMSE por Modelo')
    plt.ylabel('RMSE')
    plt.grid(True, alpha=0.3)
    
    # Panel 2: Análisis por rangos de precio
    plt.subplot(1, 2, 2)
    
    # Ordenar datos por precio real
    sorted_indices = np.argsort(y_test_orig)
    sorted_real = y_test_orig[sorted_indices]
    sorted_pred = ensemble_pred_orig[sorted_indices]
    sorted_errors = np.abs(sorted_pred - sorted_real)
    
    # Crear percentiles
    n_percentiles = 10
    percentile_size = len(sorted_real) // n_percentiles
    percentile_errors = []
    percentile_labels = []
    
    for i in range(n_percentiles):
        start_idx = i * percentile_size
        end_idx = (i + 1) * percentile_size if i < n_percentiles - 1 else len(sorted_real)
        
        percentile_errors.append(np.mean(sorted_errors[start_idx:end_idx]))
        min_val = sorted_real[start_idx]
        max_val = sorted_real[end_idx-1]
        percentile_labels.append(f"{min_val:.1f}-{max_val:.1f}")
    
    plt.bar(range(n_percentiles), percentile_errors, width=0.7)
    plt.xticks(range(n_percentiles), percentile_labels, rotation=45)
    plt.title('Error Absoluto Medio por Rango de Precio')
    plt.xlabel('Rango de Precio')
    plt.ylabel('Error Absoluto Medio')
    plt.grid(axis='y', alpha=0.3)
    
    plt.tight_layout()
    plt.savefig('model_comparison_and_errors.png', dpi=300)
    plt.show()
    
    return {
        'rmse': ensemble_rmse,
        'r2': ensemble_r2,
        'mae': ensemble_mae,
        'mape': ensemble_mape
    }

# 3. ENTRENAMIENTO DE MODELOS Y EVALUACIÓN
# =======================================

print("\n" + "="*60)
print("ENTRENANDO MODELO NEURAL NETWORK AVANZADO")
print("="*60)
start_time = time.time()

# Crear y entrenar el modelo neural mejorado
nn_model = EnhancedHousingNN(
    input_size=X_train_scaled.shape[1], 
    initial_units=320,
    dropout_rate=0.15,
    depth_factor=0.5
).to(device)

nn_model, nn_val_loss, train_losses, val_losses = train_advanced_model_v2(
    nn_model, train_loader, X_test_tensor, y_test_tensor, epochs=300, patience=30
)

training_time = time.time() - start_time
print(f"Tiempo de entrenamiento NN: {training_time:.2f} segundos")

# Evaluar modelo neural
nn_model.eval()
with torch.no_grad():
    nn_preds = nn_model(X_test_tensor).cpu().numpy()
    nn_preds_orig = scaler_y.inverse_transform(nn_preds.reshape(-1, 1)).flatten()
    y_test_orig = scaler_y.inverse_transform(y_test_scaled.reshape(-1, 1)).flatten()
    
    nn_rmse = np.sqrt(mean_squared_error(y_test_orig, nn_preds_orig))
    nn_r2 = r2_score(y_test_orig, nn_preds_orig)

print(f"NN - RMSE: {nn_rmse:.4f}, R²: {nn_r2:.4f}")

print("\n" + "="*60)
print("CREANDO ENSAMBLAJE DE MODELOS")
print("="*60)

# Entrenar y ensamblar múltiples modelos
ensemble_results = train_multi_model_ensemble(
    X_train_scaled, y_train_scaled, X_test_scaled, y_test_scaled, nn_model
)

print("\n" + "="*60)
print("EVALUACIÓN FINAL DE MODELOS")
print("="*60)

# Evaluar y visualizar resultados finales
final_metrics = evaluate_and_visualize_final_model(
    ensemble_results, y_test_scaled, scaler_y, train_losses, val_losses
)

# 4. GUARDAR MODELOS PARA USO FUTURO
# =================================
print("\n" + "="*60)
print("GUARDANDO MODELOS PARA USO FUTURO")
print("="*60)

# Guardar modelo neural
torch.save({
    'model_state_dict': nn_model.state_dict(),
    'input_size': X_train_scaled.shape[1],
    'initial_units': 320,
    'dropout_rate': 0.15,
    'depth_factor': 0.5
}, 'california_nn_model_enhanced.pth')

# Guardar modelos, escaladores y pesos del ensamblaje
with open('california_ensemble_model.pkl', 'wb') as f:
    pickle.dump({
        'models': ensemble_results['models'],
        'weights': ensemble_results['weights'],
        'scaler_X': scaler_X,
        'scaler_y': scaler_y,
        'feature_names': X_df.columns.tolist()
    }, f)

print("Modelos guardados correctamente")

# 5. FUNCIÓN PARA PREDICCIÓN USANDO MODELOS GUARDADOS
# =================================================

def predict_housing_price_ensemble(features, model_path='california_ensemble_model.pkl', nn_model_path='california_nn_model_enhanced.pth'):
    """
    Predice precio de vivienda usando el ensamblaje de modelos
    
    Args:
        features: DataFrame o array con las características originales (8 columnas)
        model_path: Ruta al archivo del modelo ensamblado
        nn_model_path: Ruta al archivo del modelo neural
    
    Returns:
        float: Precio de vivienda predicho
    """
    # Cargar modelos y escaladores
    with open(model_path, 'rb') as f:
        ensemble_data = pickle.load(f)
    
    # Extraer componentes
    models = ensemble_data['models']
    weights = ensemble_data['weights']
    scaler_X = ensemble_data['scaler_X']
    scaler_y = ensemble_data['scaler_y']
    feature_names = ensemble_data['feature_names']
    
    # Cargar modelo neural
    nn_data = torch.load(nn_model_path)
    nn_model = EnhancedHousingNN(
        input_size=nn_data['input_size'],
        initial_units=nn_data.get('initial_units', 320),
        dropout_rate=nn_data.get('dropout_rate', 0.15),
        depth_factor=nn_data.get('depth_factor', 0.5)
    ).to(device)
    nn_model.load_state_dict(nn_data['model_state_dict'])
    nn_model.eval()
    
    # Preparar características
    if isinstance(features, pd.DataFrame):
        # Asegurarse de que tiene las columnas correctas
        if set(features.columns) != set(feature_names):
            raise ValueError(f"Las columnas del DataFrame no coinciden con las esperadas: {feature_names}")
    else:
        # Convertir array a DataFrame
        if features.shape[1] == len(feature_names):
            features = pd.DataFrame(features, columns=feature_names)
        else:
            raise ValueError(f"El número de características ({features.shape[1]}) no coincide con lo esperado ({len(feature_names)})")
    
    # Aplicar feature engineering
    features_enhanced = create_enhanced_features(features)
    
    # Convertir a array
    if isinstance(features_enhanced, pd.DataFrame):
        features_array = features_enhanced.values
    else:
        features_array = features_enhanced
    
    # Escalar características
    features_scaled = scaler_X.transform(features_array)
    
    # Predicciones de cada modelo
    all_preds = []
    
    # 1. Neural Network
    features_tensor = torch.FloatTensor(features_scaled).to(device)
    with torch.no_grad():
        nn_pred = nn_model(features_tensor).cpu().numpy()
    all_preds.append(nn_pred)
    
    # 2. Gradient Boosting
    gbr_pred = models['gbr'].predict(features_scaled)
    all_preds.append(gbr_pred)
    
    # 3. Random Forest
    rf_pred = models['rf'].predict(features_scaled)
    all_preds.append(rf_pred)
    
    # 4. Extra Trees
    et_pred = models['et'].predict(features_scaled)
    all_preds.append(et_pred)
    
    # 5. XGBoost
    xgb_pred = models['xgb'].predict(features_scaled)
    all_preds.append(xgb_pred)
    
    # Combinar predicciones usando pesos
    ensemble_pred_scaled = np.zeros_like(nn_pred)
    model_names = ["Neural Network", "Gradient Boosting", "Random Forest", "Extra Trees", "XGBoost"]
    
    for i, (name, pred) in enumerate(zip(model_names, all_preds)):
        weight = weights[name]
        ensemble_pred_scaled += weight * pred
    
    # Convertir a escala original
    ensemble_pred = scaler_y.inverse_transform(ensemble_pred_scaled.reshape(-1, 1)).flatten()
    
    return ensemble_pred

# Ejemplo de uso de la función de predicción (descomentarlo para probar)
# sample_features = X_df.iloc[[0, 1, 2]]  # Tomar algunas muestras
# predictions = predict_housing_price_ensemble(sample_features)
# print("Predicciones:", predictions)

# Mostrar información de finalización
print("\n" + "="*60)
print("PROCESO COMPLETADO")
print("="*60)
print(f"Métricas finales: RMSE={final_metrics['rmse']:.4f}, R²={final_metrics['r2']:.4f}")
print("Archivos generados:")
print("  - california_nn_model_enhanced.pth (Modelo neural)")
print("  - california_ensemble_model.pkl (Ensamblaje de modelos)")
print("  - advanced_error_analysis.png (Visualización de errores)")
print("  - model_comparison_and_errors.png (Comparación de modelos)")
```
